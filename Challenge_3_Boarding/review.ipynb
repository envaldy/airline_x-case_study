```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("singapore_airlines_reviews.csv")
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
filtered_df = df[df['rating'].isin([4, 5])]

keywords = ["online check-in","online check in","kiosk","mobile check"]

def contains_keywords(text):
    return any(keyword in text.lower() for keyword in keywords)

good_online_boarding_df = filtered_df[filtered_df['text'].apply(contains_keywords)]
count = good_online_boarding_df.shape[0]
print("Number of times passengers gave a rating of 4 or 5 and mentioned good online boarding:", count)
```

```python
good_online_boarding_df
```

```python
filtered_df2 = df[df['rating'].isin([1, 2])]

keywords = ["online check-in","online check in","kiosk","mobile check"]

def contains_keywords(text):
    return any(keyword in text.lower() for keyword in keywords)

bad_online_df = filtered_df2[filtered_df2['text'].apply(contains_keywords)]
count_bad = bad_online_df.shape[0]
print("Number of times passengers gave a rating of 4 or 5 and mentioned good online boarding:", count_bad)
```

```python
counts = {
    'category': ['good', 'bad review on online check-in'],
    'weight': [40, 12]
}

df_count = pd.DataFrame(counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_count
```

```python
plt.figure(figsize=(8, 6))
plt.pie(df_count['weight'], labels=df_count['category'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('online check-in toward satisfaction')
plt.show()
```
