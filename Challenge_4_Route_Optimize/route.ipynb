```python
df_o = pd.read_csv("case_route.csv")
```

```python
df_o
```

```python
df_o.drop(columns=["Unnamed: 0","Revenue per Flight","Operating Cost","Net Revenue"],inplace=True)
```

```python
df_o["times"] = [3, 10, 7, 2, 0]
```

```python
df_o
```

```python
plt.figure(figsize=(10,8))
df_sorted = df_o.sort_values(by='times', ascending=False)
barplot = sns.barplot(y='Flight Revenue',x='times',data=df_sorted,errorbar=None, palette='viridis')

plt.ylabel('Flight')

for p in barplot.patches:
    barplot.annotate(format(p.get_width(), '.0f'),
                     (p.get_width(), p.get_y() + p.get_height() / 2.),
                     ha = 'center', va = 'center',
                     xytext = (15, 0),
                     textcoords = 'offset points')
plt.title('Flight Frequent')
plt.xticks(rotation=0)
```
